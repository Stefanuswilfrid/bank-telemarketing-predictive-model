



# **Business Problem**
## **Portugal Banking Campaign Opportunity**

The banking industry in Europe, particularly in Portugal, has undergone significant transformation following the financial crisis and pandemic. One marketing approach that remains relevant and highly impactful is direct marketing campaigns through telephone (telemarketing) for term deposit products. Although this method is considered conventional, it continues to demonstrate high effectiveness in reaching potential customers, especially for financial products that require detailed explanation.

According to the European Banking Federation (2023) report, personalization strategy and utilization of customer historical data are key factors in improving the success of marketing campaigns in the financial sector. Data from Portugal bank marketing campaigns collected by the UCI Machine Learning Repository shows that although most customers were contacted by phone, only about 11.3% eventually agreed to open a term deposit account.

This fact indicates a need to understand more deeply what factors influence customer decisions. By utilizing data analysis techniques and predictive models, financial institutions can improve campaign effectiveness, reduce marketing costs, and increase conversion rates.

## **Term Deposit Campaign Problem**

One of the main challenges in marketing term deposit products in the banking sector is the low customer conversion rate. Many customers reject offers, even though they fall within the relevant target market. This is likely due to marketing approaches that are not personal enough, inappropriate timing in contacting customers, or lack of information about customer preferences and needs.

A study by Gensler et al. (2012) shows that marketing campaign success heavily depends on the ability to segment customers and deliver messages at the right time. Inefficiency in targeting high-potential customers leads to high operational costs and low return on investment (ROI) from campaigns.

## **Problem Statement**

The bank faces challenges in determining which potential customers have a high likelihood of accepting term deposit offers. Inaccurate campaign targeting results in resource waste and decreased marketing campaign effectiveness.

## **Goals**

Based on these problems, the main objective of this analysis is to develop a classification model that can predict the likelihood of a customer accepting a term deposit offer. This model will be based on demographic data, employment status, previous marketing interaction history, and other variables available in the dataset.

With this model, the marketing team can focus more on targeting customers who have higher potential for conversion, thereby increasing campaign efficiency and improving the success rate of banking product marketing.

## **Matrix Evaluation**

### Conversion Rate Calculation

Conversion rate is a key performance indicator (KPI) in marketing campaigns that shows the percentage of customers who take the desired action, in this case, `subscribing to a term deposit` compared to the total number of customers contacted.
$$
\text{Conversion Rate} = \left( \frac{\text{Number of customers who subscribed}}{\text{Total customers contacted}} \right) \times 100\%
$$
Actual campaign data from Portugal (UCI Machine Learning Repository) shows that of all customers contacted, only about 11.27% agreed to open a term deposit account.   
Although this figure can be said to reflect moderate results, this conversion still lags when compared to higher-performing financial industry performance. According to Ruler Analytics (2023) report, the average conversion rate in high-performing financial campaigns can reach 23%.

This gap indicates significant opportunity to improve telemarketing strategy effectiveness through a more precise data-driven approach.

### ROMI Calculation

Return on Marketing Investment (ROMI) is an important metric to measure how much profit is generated from marketing investments. ROMI becomes a strategic tool for assessing campaign effectiveness, planning future budget allocation, and comparing performance across marketing channels.
$$
\text{ROMI} = \left( \frac{\text{Revenue generated} - \text{Marketing costs}}{\text{Marketing costs}} \right) \times 100\%
$$
Calculation results show that Electric Bank's current ROMI stands at 144.92%. This means that for every 1 euro spent in telemarketing campaigns, the bank generates 1.45 euros in net profit.
However, this figure is still below industry standards, where ROMI considered "good" for the financial industry is 5:1 (500%) or more, according to a report by Improvado (2023).

This data shows significant room to improve profitability from marketing activities, especially by optimizing customer targeting and minimizing operational cost waste through smarter analytical approaches.

### Cost Analysis

Before building the machine learning model, initial business analysis was conducted to understand the financial impact of each prediction decision in the telemarketing campaign.

- Cost and Revenue Assumptions.
    Based on study by Moro, Cortez, & Rita (2014):

    - Cost of one telemarketing call: 23 euros
    - Revenue from one successfully subscribed customer: 500 euros


- Cost and Profit Simulation Based on Prediction Types.

    | Prediction Type        | Explanation                                                             | Net Gain  |
    | ----------------------| ----------------------------------------------------------------------- | --------- |
    | **True Positive (TP)**| Customer predicted to subscribe, and actually does.                      | +477 euros|
    | **False Positive (FP)**| Customer predicted to subscribe, but doesn't.                           | −23 euros |
    | **True Negative (TN)**| Customer predicted not to subscribe, and indeed doesn't.                 | 0 euros   |
    | **False Negative (FN)**| Customer predicted not to subscribe, but should have.                   | 0 euros   |


- Chosen Evaluation Metrics.

    - Precision:
        - Because each False Positive causes direct loss in the form of telemarketing costs that don't generate revenue, Precision is the most appropriate evaluation metric
        - With high precision, we ensure that predictions of potentially subscribing customers are truly accurate, so marketing costs aren't wasted and ROI increases.
