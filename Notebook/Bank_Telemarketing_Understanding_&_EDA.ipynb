# Data Understanding

## Bank Marketing Campaign Dataset Description
This dataset contains information about direct marketing campaigns (through phone calls) of a banking institution. The goal is to predict whether a client will subscribe to a term deposit after the marketing campaign.

The dataset is divided into several attribute groups for easier understanding.

## 1. Client Data
*Basic information about the client.*

### age
- **Data Type:** Numeric
- **Description:** Client's age in years

### job
- **Data Type:** Categorical
- **Categories:** "admin.", "blue-collar", "entrepreneur", "housemaid", "management", "retired", "self-employed", "services", "student", "technician", "unemployed", "unknown"
- **Description:** Type of client's job

### marital
- **Data Type:** Categorical
- **Categories:** "divorced", "married", "single", "unknown"
- **Note:** "divorced" includes widowed

### education
- **Data Type:** Categorical
- **Categories:** 
  - "basic.4y" (4 years basic)
  - "basic.6y" (6 years basic)
  - "basic.9y" (9 years basic)
  - "high.school"
  - "illiterate"
  - "professional.course"
  - "university.degree"
  - "unknown"

### default
- **Data Type:** Categorical
- **Categories:** "no", "yes", "unknown"
- **Description:** Does the client have credit in default?

### housing
- **Data Type:** Categorical
- **Categories:** "no", "yes", "unknown"
- **Description:** Does the client have a housing loan?

### loan
- **Data Type:** Categorical
- **Categories:** "no", "yes", "unknown"
- **Description:** Does the client have a personal loan?

## 2. Data Related to the Last Contact of the Current Campaign
*Information related to the last contact made in the current marketing campaign.*

### contact
- **Data Type:** Categorical
- **Categories:** "cellular", "telephone"
- **Description:** Type of communication medium used for contact

### month
- **Data Type:** Categorical
- **Categories:** "jan", "feb", "mar", ..., "nov", "dec"
- **Description:** Last contact month

### day_of_week
- **Data Type:** Categorical
- **Categories:** "mon", "tue", "wed", "thu", "fri"
- **Description:** Day of the week when the last contact was made

### duration
- **Data Type:** Numeric (seconds)
- **Description:** Duration of the last contact in seconds
- **Important Note:** This attribute highly influences the target variable (y). If duration=0, y is always "no"

## 3. Other Attributes
*Other information relevant to the campaign.*

### campaign
- **Data Type:** Numeric
- **Description:** Number of contacts performed during this campaign for this client (including last contact)

### pdays
- **Data Type:** Numeric
- **Description:** Number of days that passed since the client was last contacted from a previous campaign
- **Note:** Value of 999 means client was not previously contacted

### previous
- **Data Type:** Numeric
- **Description:** Number of contacts performed before this campaign for this client

### poutcome
- **Data Type:** Categorical
- **Categories:** "failure", "nonexistent", "success"
- **Description:** Outcome of the previous marketing campaign

## 4. Social and Economic Context Attributes
*These variables reflect social and economic conditions at the time of contact.*

### emp.var.rate
- **Data Type:** Numeric
- **Description:** Employment variation rate - quarterly indicator

### cons.price.idx
- **Data Type:** Numeric
- **Description:** Consumer price index - monthly indicator

### cons.conf.idx
- **Data Type:** Numeric
- **Description:** Consumer confidence index - monthly indicator

### euribor3m
- **Data Type:** Numeric
- **Description:** Euribor 3 month rate - daily indicator

### nr.employed
- **Data Type:** Numeric
- **Description:** Number of employees - quarterly indicator

## 5. Target Variable (Output Variable)
*This is the variable you want to predict.*

### y
- **Data Type:** Binary
- **Categories:** "yes", "no"
- **Description:** Has the client subscribed to a term deposit?

```python
#Correlation Between Numeric Features
numeric_cols = ['age', 'campaign', 'previous',
                'emp.var.rate', 'cons.price.idx', 'cons.conf.idx',
                'nr.employed']

corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Numeric Features')
plt.show()
```


# Correlation Analysis

## Strong Positive Correlations
- emp.var.rate - nr.employed: 0.90
- emp.var.rate - cons.price.idx: 0.77
- This indicates that the employment variation rate is strongly related to the number of employed people and the consumer price index.
- The higher the employment rate, the higher the number of people employed and consumer goods prices (mild inflation).

## Significant Negative Correlations
- previous - nr.employed: â€“0.49
- previous - emp.var.rate: â€“0.40
- Meaning: the higher the number of previous contacts, the lower the employment rate and economic variables.
- This likely indicates that marketing efforts were more aggressive during economic downturns.

## Weak Correlations
- Most other variables such as age, campaign, and cons.conf.idx have weak correlations with other variables.
- These variables might still be relevant in non-linear ways or in interactions with categorical variables.
- Variables like campaign and pdays have very weak correlations with other features. This means these factors tend to be linearly independent from other numeric features.


# Actionable Steps

## Monitor Over-contacting (previous)
- Negative correlations indicate that too frequent contact with prospects in the past could have adverse effects.
- Set limits on the number of repeat contacts within a period. Create segmentation based on previous responses.

## Monitor Consumer Price Index (cons.price.idx)
- If consumer prices increase, customers are likely to become more selective in spending money.
- Modify offering strategies (discounts or special offers).

## Avoid Generalizing from Weak Correlations
- Features like age, campaign, and cons.conf.idx can still be important through:
  - Interactions with other features (for example: age and job type)
  - Non-linear models such as decision trees or ensemble models
